```python
import pandas as pd
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.charts import Area, Horizon, Scatter
from bokeh.palettes import PuBuGn, Category20
```

```python
df = pd.read_csv('bachelors.csv', index_col='Year')
```

```python
df.head()
```

```python
p = figure()

p.circle(x=df.index, y=df['Architecture'])
output_notebook()
show(p)
```

```python
area = Area(df[['Biology','Architecture', 'Agriculture']], title="Area Chart", legend="top_left",
            xlabel='time', ylabel='memory')
output_notebook()
show(area)
```

```python
hor = Horizon(df[['Biology','Architecture', 'Agriculture']], title="Area Chart", legend="top_left",
            xlabel='time', ylabel='memory')

show(hor)
```

```python
scat = Scatter(df[['Biology','Architecture', 'Agriculture']], title="Area Chart", legend="top_left",
            xlabel='time', ylabel='memory')

show(scat)
```
